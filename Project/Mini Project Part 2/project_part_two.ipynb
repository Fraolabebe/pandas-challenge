```python
# Dependencies
import pandas as pd 
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
coders_df=pd.read_csv(file)
coders_df
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
Extracted_coders_df= coders_df.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
Extracted_coders_df
```

```python
Extracted_coders_df.columns
```

```python
print("This is number of data types by column of the extracted data frame")
Extracted_coders_type = Extracted_coders_df.dtypes
Extracted_coders_type.value_counts()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
print("Changed the the response of attendees 0 to NO 1 to YES in the data frame")

Extracted_coders_df['AttendedBootcamp'] = Extracted_coders_df['AttendedBootcamp'].replace({0: "No", 1: "Yes"})
Extracted_coders_df.head()


```

```python
#Calculate the total number of respondents in survey
total_surveyed = Extracted_coders_df["AttendedBootcamp"].count()
total_surveyed

# calculate the number of respondents who attended bootcamp
Extracted_coders_df["AttendedBootcamp"].value_counts()
```

```python
# Extract rows for only those who attended a bootcamp
attended_bootcamp = Extracted_coders_df.loc[Extracted_coders_df["AttendedBootcamp"] == "Yes"]
attended_bootcamp.head()
```

```python
attended_bootcamp.describe()
```

```python
# Calculate average age of attendees
age_attended_avg =attended_bootcamp["Age"].mean()

# Calculate how many people attended a bootcamp
total_attended = attended_bootcamp["AttendedBootcamp"].count()

# Calculate how many attendees hold degrees
school_degrees = attended_bootcamp["SchoolDegree"].loc[
    (attended_bootcamp["SchoolDegree"] != "some college credit, no degree") &
    (attended_bootcamp["SchoolDegree"] != "high school diploma or equivalent (GED)") &
    (attended_bootcamp["SchoolDegree"] != "trade, technical, or vocational training") &
    (attended_bootcamp["SchoolDegree"] != "some high school") &
    (attended_bootcamp["SchoolDegree"] != "no high school (secondary school)")].count()

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification


# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
total_gender_df = attended_bootcamp["Gender"].value_counts()
total_gender = attended_bootcamp["Gender"].count()
male = attended_bootcamp["Gender"].value_counts()["male"]
female = attended_bootcamp["Gender"].value_counts()["female"]
non_gender_specific = total_gender - male - female
print(f" Total: {total_gender}\n Male: {male}\n Female: {female}\n non_specific: {non_gender_specific}")

# Calculate percentage of respondents who attended a bootcamp
percent_attended = (total_attended/total_surveyed)*100

# Calculate percentage of respondents belonging to each gender
male_percent = (male/total_gender)*100
female_percent = (female/total_gender)*100   
non_gender_specific = (non_gender_specific/total_gender)*100
print(f" Male % is {male_percent},\n Female % is {female_percent},\n and Non gender specific % is {non_gender_specific}")

# Calculate percentage of attendees with a school degree
holds_degree_percentage = (school_degrees/total_attended)*100
holds_degree_percentage

# Calculate average post-bootcamp salary
post_bootcamp_avg_salary = (attended_bootcamp["BootcampPostSalary"].mean())


```

```python
bootcamp_table = pd.DataFrame ({"Total Respondents": [total_surveyed],
                                "Total Bootcamp Attendees": [total_attended],
                                "Percent Attended Bootcamp": [percent_attended],
                                "Average Age": [age_attended_avg],
                                "Percent Male": [male_percent],
                                "Percent Female": [female_percent],
                                "Percent Non Specific": [non_gender_specific],
                                "Percent with a Degree": [holds_degree_percentage],
                                "Average Post Bootcamp Salary": [post_bootcamp_avg_salary]})
bootcamp_table = bootcamp_table[["Total Respondents",
                                "Total Bootcamp Attendees",
                                "Percent Attended Bootcamp",
                                "Average Age",
                                "Percent Male",
                                "Percent Female",
                                "Percent Non Specific",
                                "Percent with a Degree",
                                "Average Post Bootcamp Salary"]]
bootcamp_table = bootcamp_table.round(2)



bootcamp_table
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
Bootcampnames_df= pd.DataFrame(coders_df["BootcampName"].value_counts())
Bootcampnames_df
Bootcampnames_df.reset_index(inplace=True)
Bootcampnames_df.columns = ["BootcampName", "Count"]
Bootcampnames_df.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommend_attendees = coders_df.replace({"yes": 1, "No": 0})
recommend_attendees = pd.DataFrame(recommend_attendees.groupby("BootcampName")["BootcampRecommend"].sum())
recommend_attendees.reset_index(inplace=True)
recommend_attendees.columns=["BootcampName", "Recommenders"]
recommend_attendees.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
combined_df = pd.merge(Bootcampnames_df, recommend_attendees, on="BootcampName")
combined_df.head()
```

```python
highly_recommend_bootcamp = recommend_bootcamp["BootcampRecomment"]
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
Recommenders_count = combined_df["Recommenders"].count()
Recommenders_count

combined_df["Percentage of Recommended"] = (combined_df["Recommenders"]/Recommenders_count)*100
combined_df = combined_df.sort_values(["Percentage of Recommended"], ascending=False).round(2)
combined_df["Percentage of Recommended"] = combined_df["Percentage of Recommended"].map("{0:,.1f}%".format)
combined_df.head()
```
